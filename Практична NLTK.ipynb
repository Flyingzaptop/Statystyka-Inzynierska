Wydaj token tekstu:
 * Podziel tekst na zdania
 * Podziel każde zdanie na słowa. Wynik powinien być „lista [lista]”, w której mówimy „dane [5] [10]” oznacza 10. słowo z 5. zdania (indeksowanie z 0)
 * Usuń interpunkcję i strzałe słowo
 * Utwórz listę wszystkich słów z tekstu „słowa”

```python
text = """Mrs Pig was very tired: 'Oh dear,' she said to her three little pigs, 'I can’t do this work anymore, I’m afraid you must leave home and make your own way in the world.' So the three little pigs set off.

The first little pig met a man carrying a bundle of straw.

'Excuse me,' said the first little pig politely. 'Would you please sell some of your straw so I can make a house?'

The man readily agreed and the first little pig went off to find a good place to build his house.

The other little pigs carried on along the road and, soon, they met a man carrying a bundle of sticks.

'Excuse me,' said the little pig politely. 'Would you please sell me some sticks so I can build a house?'

The man readily agreed and the little pig said goodbye to his brother.

The third little pig didn’t think much of their ideas:

'I’m going to build myself a much bigger, better, stronger house,' he thought, and he carried off down the road until he met a man with a cart load of bricks.

'Excuse me,' said the third little pig, as politely as his mother had taught him. 'Please can you sell me some bricks so I can build a house?'

'Of course,' said the man. 'Where would you like me to unload them?'

The third little pig looked around and saw a nice patch of ground under a tree.

'Over there,' he pointed.

They all set to work and by nighttime the house of straw and the house of sticks were built but the house of bricks was only just beginning to rise above the ground. The first and second little pigs laughed, they thought their brother was really silly having to work so hard when they had finished."""
```

```python
!pip install nltk
```

```python
import nltk
import string
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
sentences = nltk.sent_tokenize(text)
words = []
stop_words = set(stopwords.words('english'))
for sentence in sentences:
    sentence_words = []
    for word in nltk.word_tokenize(sentence):
      word = word.lower()
      if word not in string.punctuation and word not in stop_words:
            sentence_words.append(word)
    words.append(sentence_words)

all_words = []
for sentence_words in words:
  all_words.extend(sentence_words)
```

# Zadanie 2

 * Wykonuj słowa
 * Wykonaj latymizację słów
 * Utwórz `pandas.dataframe` ze słowami, które mają różne wyniki w zakresie zbierania i lemmatyzacji. Powinny być 3 kolumn:
  * Oryginalna sowa
  * Wynik schematu
  * Wynik cytrynowania

```python
import pandas as pd
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_words = [stemmer.stem(word) for word in all_words]
lemmatized_words = [lemmatizer.lemmatize(word) for word in all_words]

diff_words = []
for i in range(len(all_words)):
  if stemmed_words[i] != lemmatized_words[i]:
    diff_words.append([all_words[i], stemmed_words[i], lemmatized_words[i]])

df = pd.DataFrame(diff_words, columns=['Оригінальне слово', 'Результат стемінгу', 'Результат лематизації'])
df
```

Podkreśl frazy następującymi szablonami:
  * Przymiotnik + rzeczownik
  * (co najmniej 1 przymiotnik) + rzeczownik
  * (przysłówek + czasownik) lub (czasownik + przysłówek)

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentences = sent_tokenize(text)
phrases = []

for sentence in sentences:
    tokens = word_tokenize(sentence)
    tagged_tokens = pos_tag(tokens)

    # Шаблон 1 та 2: прикметник + іменник (хоча б 1 прикметник) + іменник
    for i in range(len(tagged_tokens) - 1):
        if tagged_tokens[i][1].startswith('JJ') and tagged_tokens[i+1][1].startswith('NN'):
            phrases.append(" ".join([tagged_tokens[i][0], tagged_tokens[i+1][0]]))

    # Шаблон 3: (прислівник + дієслово) АБО (дієслово + прислівник)
    for i in range(len(tagged_tokens) - 1):
      if (tagged_tokens[i][1].startswith('RB') and tagged_tokens[i+1][1].startswith('VB')) or \
         (tagged_tokens[i][1].startswith('VB') and tagged_tokens[i+1][1].startswith('RB')):
        phrases.append(" ".join([tagged_tokens[i][0], tagged_tokens[i+1][0]]))

phrases
```
