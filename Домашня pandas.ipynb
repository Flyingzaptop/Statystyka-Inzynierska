# Zadanie 6
Sprawdź, czy całkowity koszt zakupu jest odpowiednio obliczony. Wydaj te linie, w których występuje błąd. Prawidłowe dane w tabeli

```python
df['sale_amt_calculated'] = df['units'] * df['unit_price']
df.loc[df['sale_amt'] != df['sale_amt_calculated'], 'sale_amt'] = df['sale_amt_calculated']
df = df.drop('sale_amt_calculated', axis=1)
df
```
